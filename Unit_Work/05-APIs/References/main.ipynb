```python
import pandas as pd
import random
from stock import get_stock
```

```python
stocklist = ["MSFT","AAPL", "GOOG", "ENPH", "CAT", "XOM", "DIS"]
stocks = get_stock(stocklist)
closes = stocks.xs("close", axis="columns", level=1)
closes
```

```python
stocks["XOM"]
```

```python
returns = closes.pct_change()
returns
```

```python
preweight = [random.random() for w in range(len(stocklist))]
weightsum = sum(preweight)
weights = [w/weightsum for w in preweight]
```

```python
weights
```

```python
initial = 10000
portfolio_returns = returns.dot(weights)
cumulative_returns = (1+portfolio_returns).cumprod()
(initial*cumulative_returns).plot()
```

```python
results = []

trials = 2000

for run in range(trials):
    

    # generate weights
    preweight = [random.random() for w in range(len(stocklist))]
    weightsum = sum(preweight)
    weights = [w/weightsum for w in preweight]

    # calc cumulative returns
    initial = 10000
    portfolio_returns = returns.dot(weights)
    cumulative_returns = (1+portfolio_returns).cumprod()

    risk = portfolio_returns.std()
    final_return = cumulative_returns[-1]

    run_result = {
        "weights": weights,
        "total_return": final_return,
        "risk": risk,
        "sharpe": final_return/risk
    }

    results.append(run_result)
```

```python
results_df = pd.DataFrame(results)
results_df
```

```python
results_df = results_df.sort_values("sharpe", ascending=False)
```

```python
results_df.plot(kind="scatter", x="risk", y="total_return", c="sharpe")
```
